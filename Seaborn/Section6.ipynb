```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
sns.kdeplot(df.price,shade=True); #Yogunluk grafiginin altini doldurarak gosterir.
```

```python
(sns
.FacetGrid(df,
                hue="cut",
                height=5,
                xlim=(0,10000),
                )
.map(sns.kdeplot,"price",shade=True)
.add_legend()
);
```

```python
sns.catplot(x="cut",y="price",hue="color",kind="point",data=df);
```
