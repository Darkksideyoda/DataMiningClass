```python
#eksik veride töhmet uygulamalari
from sklearn.preprocessing import Imputer

#Ortalama töhmet (mean imputation)
imp = Imputer(missing_values=np.nan, strategy='mean')
ortalama_tohmet =pd.DataFrame( imp.fit_transform(trainDS.iloc[:,0:5]), columns=trainDS.columns.values.tolist()[0:5])
print(ortalama_tohmet)

#Medyan tohmet
imp = Imputer(missing_values=np.nan, strategy='median')
medyan_tohmet =pd.DataFrame( imp.fit_transform(trainDS.iloc[:,0:5]), 
                            columns=trainDS.columns.values.tolist()[0:5])
print(medyan_tohmet)


#En fazla olan değer
imp = Imputer(missing_values=np.nan, strategy='most_frequent')
en_cok_tohmet =pd.DataFrame( imp.fit_transform(trainDS.iloc[:,0:5]), columns=trainDS.columns.values.tolist()[0:5])
print(en_cok_tohmet)
```
