```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
figure_size=[4,2]
```

```python
x=np.linspace(0, np.pi*2)
y=np.sin(x)
"""print(x)
print(y)"""
```

```python
plt.figure(figsize=figure_size)
plt.plot(x,y)
plt.ylabel("y değişkeni")
plt.xlabel("x değişkeni")
plt.title("Örnek")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=figure_size)
plt.plot(x,y)

plt.plot(x,y*0.3, "--", color="red", linewidth=5)
plt.legend(["1.çizgi", "2.çizgi"])

plt.ylabel("y değişkeni")
plt.xlabel("x değişkeni")
plt.title("Örnek")


plt.grid()
plt.show()
```

```python
plt.figure(figsize=[10,6])

plt.plot(x,y,'o')

plt.ylabel("y değişkeni")
plt.xlabel("x değişkeni")
plt.title("Örnek")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=[10,6])
y=[10,5,3]
x_labels=['ocak','şubat','mart']
x_pos=np.arange(len(y))
#print(x, y, x_pos, x_labels)
plt.bar(x_pos, y, color=['black','red','blue'], 
        edgecolor=['r','b','y'],
        linewidth=5)
plt.xlabel("aylar")
plt.ylabel("satış")
plt.xticks(x_pos, x_labels)
plt.title("satış")
plt.show()
```
